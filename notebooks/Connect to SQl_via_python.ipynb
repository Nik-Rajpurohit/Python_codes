```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
server = r"NIKHIL\SQLEXPRESS"
database = "inventoryDB"

engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)
```

```python
table_list = pd.read_sql_query(
    "SELECT TABLE_NAME FROM INFORMATION_SCHEMA.TABLES WHERE TABLE_TYPE='BASE TABLE'",
    con=engine
)
print(table_list)
```

```python
for table in table_list['TABLE_NAME']:
   print('-'*50, f'{table}','-'*50)
   print('COunt of records:', pd.read_sql(f'select count(*) as counts from {table}', con=engine)['counts'].values[0])
   display(pd.read_sql(f'select top 5 *  from {table}',con=engine))
```
