```python
from sqlalchemy import create_engine
import pandas as pd
import os
import time
import logging

logging.basicConfig(
    filename=r'C:\Users\nikhi\Downloads\logs\ingestion_db.log',
    level=logging.DEBUG,
    format="%(asctime)s- %(levelname)s- %(message)s",
    filemode="a"
)

server = r'NIKHIL\SQLEXPRESS'         
database = 'inventoryDB'  

engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)

def ingest_db(df, table_name, engine):
    df.to_sql(table_name,con=engine, if_exists='replace', index=False)

def load_raw_data():
    start=time.time()
    for file in os.listdir(r'C:\Users\nikhi\Downloads\data'):
        if '.csv' in file:
         df=pd.read_csv(r'C:\Users\nikhi\Downloads\data/'+file)
         logging.info(f'Ingesting {file} in db')
         ingest_db(df, file[:-4], engine)
    end=time.time()
    total_time= (end-start)/60
    logging.info('-----------Ingestion Complete--------------')
    logging.info(f'\nTotal time taken min: {total_time}')

if __name__ == "__main__":
    load_raw_data()
```
